```python
# import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9.0, 7.0)
```

```python
dataset = pd.read_csv("./slr06.csv")
```

```python
X = dataset.X
Y = dataset.Y
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
b0 = 0
b1 = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = b1*X + b0  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    b1 = b1 - L * D_m  # Update m
    b0 = b0 - L * D_c  # Update c
    
print (b1, b0)
```

```python
def predict(test):
    predictions= list()
    yhat = b0 + b1*test.X
    predictions.append(yhat)
    return predictions
```

```python
pred = predict(dataset)
pred = np.array(pred)
pred = pred.transpose()
```

```python
plt.scatter(X, Y) 
plt.plot(dataset.X, pred, color='red')  # regression line
plt.show()
```
